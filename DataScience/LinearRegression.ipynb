## Importing the Relevant Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
#Overriding the default matplotlib styles with seaborn
import seaborn as sns
sns.set()
from sklearn.datasets import load_boston
boston = load_boston()
```

## Find out Data Dimension

```python
boston.data.shape
```

## Find out feature names

```python
boston.feature_names
```

## Converting dataset into dataframe

```python
data = pd.DataFrame(boston.data)
data.columns = boston.feature_names
 
data.head(10)
```

## Adding 'Price'(target) column to the data

```python
boston.target.shape
data['Price'] = boston.target
data.head()
```

## Description of the dataset

```python
data.describe()
```

## Info of the dataset

```python
data.info()
```

```python
x = boston.data

y = boston.target

from sklearn.model_selection import train_test_split
 
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size =0.2,
                                                    random_state = 0)
  
print("xtrain shape : ", xtrain.shape)
print("xtest shape  : ", xtest.shape)
print("ytrain shape : ", ytrain.shape)
print("ytest shape  : ", ytest.shape)
```

## Applying Linear Regression Model to the dataset

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(xtrain, ytrain)
  
y_pred = regressor.predict(xtest)
```

## Plotting Scatter graph to show the prediction results – ‘ytrue’ value vs ‘y_pred’ value 

```python
plt.scatter(ytest, y_pred, c = 'green')
plt.xlabel("Price: in $1000's")
plt.ylabel("Predicted value")
plt.title("True value vs predicted value : Linear Regression")
plt.show()
```

```python
x = sm.add_constant(ytest)
results =sm.OLS(y_pred,ytest).fit()
results.summary()
```

```python
yhat = 0.9350*ytest
fig = plt.plot(ytest,yhat,c='orange',label='regression line')
plt.scatter(ytest, y_pred, c = 'green')
plt.xlabel("Price: in $1000's")
plt.ylabel("Predicted value")
plt.title("True value vs predicted value : Linear Regression")
plt.show()
```

## Results of Linear Regression.

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(ytest, y_pred)
print("Mean Square Error : ", mse)
```
